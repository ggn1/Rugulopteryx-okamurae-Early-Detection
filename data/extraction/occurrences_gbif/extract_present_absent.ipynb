# Extract Present Absent

This file contains code that separates present data from absent data and then drops duplicates.

```python
# IMPORTS
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# DATA PATHS
dir_root = "../../db/occurrence"
dir_src = f"{dir_root}/raw"
dir_dst = f"{dir_root}/processed"
```

```python
# LOAD DATA
data = pd.read_csv(f"{dir_src}/gbif_0013072-251025141854904.csv", sep="\t")
data = data.dropna(subset=["decimalLatitude", "decimalLongitude", 
                           "year", "occurrenceStatus"])
```

```python
# PROCESS DATA
data = data[["decimalLatitude", "decimalLongitude", "year", "occurrenceStatus"]]
# Separate into present and absent data. Drop duplicates.
data_present = data[data.occurrenceStatus == "PRESENT"].drop_duplicates()
data_absent = data[data.occurrenceStatus == "ABSENT"].drop_duplicates()
# Drop unnecessary column.
data_present.drop(columns=["occurrenceStatus"], axis=1, inplace=True)
data_absent.drop(columns=["occurrenceStatus"], axis=1, inplace=True)
# Drop rows with missing year data.
data_present.dropna(subset=["year"], inplace=True)
data_absent.dropna(subset=["year"], inplace=True)
# Reset indices.
data_present.reset_index(drop=True, inplace=True)
data_absent.reset_index(drop=True, inplace=True)
```

```python
# VIEW ALL DATA
px.scatter_geo(data.dropna(subset="year"), 
               lat="decimalLatitude", lon="decimalLongitude", 
               hover_data=["year"], title="Presence & Absence Records")
```

```python
# VIEW PRESENCE DATA
px.scatter_geo(data_present, lat="decimalLatitude", lon="decimalLongitude", 
               hover_data=["year"], title="Presence Records")
```

```python
# VIEW ABSENCE DATA
px.scatter_geo(data_absent, lat="decimalLatitude", lon="decimalLongitude", 
               hover_data=["year"], title="Absence Records")
```

```python
# VIEW YEAR DISTRIBUTION
print("Year Distribution")
print(data.year.describe())
print("\nTop 5 years with most presence records:")
print(data_present['year'].value_counts().head(5))
print("\nTop 5 years with most absence records:")
print(data_absent['year'].value_counts().head(5))
```
